```python
import pandas as pd

df = pd.read_csv('titanic.csv')
df
```

```python
#PIVOT TABLES
#1.AVERAGE SURVIVAL BY GENDER
df.pivot_table(index='sex', values='survived', aggfunc='mean')
```

```python
#2.COUNT BY PASSENGERS BY CLASS
df.pivot_table(index='pclass', values='name', aggfunc='count')
```

```python
#3.AVERAGE AGE BY PASSENGER CLASS
df.pivot_table(index='pclass',values='age', aggfunc='mean')
```

```python
#4.COUNT SURVIVAL BY CLASS AND GENDER
df.pivot_table(values = 'survived', index = ['pclass', 'sex'], aggfunc = ['count'])
```

```python
#5.FAIR PAID BY CLASS (AVERAGE)
df.pivot_table(values='fare',index='pclass',aggfunc='mean')
```

```python
#6. NUMBER OF PASSENGERS BY EMBARKATION PORT
df.pivot_table(values='name',index='embarked',aggfunc='count')
```

```python
#7. MEAN OF SURVIVAL BY EMBARKATION PORT
df.pivot_table(values='survived',index='embarked',aggfunc='mean')
```

```python
#8. AVERAGE AGE BY GENDER
df.pivot_table(values='age',index='sex',aggfunc='mean')
```

```python
#9. SURVIVAL BY CLASS (COUNT AND MEAN)
df.pivot_table(values='survived',index='pclass',aggfunc=['count','mean'])
```

```python
#10. GET MIN,MAX,COUNT OF PCLASS AND FARE COLUMN CONSIDERING GENDER AND EMBARK AS INDEX
df.pivot_table(values=['pclass','fare'],index=['sex','embarked'],aggfunc={'pclass':'count','fare':['min','max']})
```
