```python
import pandas as pd

df = pd.read_csv('titanic.csv')
display(df.head())
```

```python
df
```

```python
#1.WAQ to retrieve all passengers who embarked from southampton(embarked == 'S')

df[df['embarked'] == 'S']
```

```python
#2.WAQ to find passengers who paid a fare between 20 and 40 units of currency

df[(df['fare'] > 20) & (df['fare'] < 40)]
```

```python
#3. WAQ to find passengers whose age is missing(age == NaN)

df[df['age'].isna()]
```

```python
#4.WAQ to retrieve the passengers who had siblings or spouses aboard the titanic

df[df['sibsp'] > 0]
```

```python
#5. WAQ to find passengers who survived and paid more than median fare

df[(df['survived'] == 1) & (df['fare'] > df['fare'].median())]
```

```python
df['age'] = df['age'].fillna(df['age'].median()).round().astype(int)
```

```python
#6. WAQ to find the passengers who were under the age of 18

df[df['age'] < 18]
```

```python
#7. WAQ to retrieve only the passengers who were both female and survived

df[(df['sex'] == 'female') & (df['survived'] == 1)]
```

```python
#8. WAQ to retrive the passengers who travelled in either first or second class

df[(df['pclass'] == 1) | (df['pclass'] == 2)]
```

```python
#9. WAQ to find children (passengers under 18) who survived the disaster

df[(df['age'] < 18) & (df['survived'] == 1)]
```

```python
#10. WAQ to find all female passengers who travelled in first class

df[(df['sex'] == 'female') & (df['pclass'] == 1)]
```
