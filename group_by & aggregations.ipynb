```python
#QUESTIONS ON GROUP BY AND DATA AGGREGATE FUNCTIONS
import pandas as pd

df = pd.read_csv('amazon_fires.csv', encoding='latin1')
df
```

```python
#1. NO OF EMPLOYEES WHO GOT FIRED IN THE MONTH OF JANUARY
df.groupby('mes').size().loc['Janeiro']
```

```python
#2. NO OF PEOPLE WHO GOT FIRED IN 2002(NEED FEATURE ENGINEERING)
df['encontro']
df['year'] = pd.to_datetime(df['encontro']).dt.year
df.groupby('year').size().loc[2002]
```

```python
#3. NO OF FIRES IN YEAR 2001 IN TOCANTINS
df.groupby(['estado', 'year'])['numero'].sum().loc[('Tocantins', 2001)]
```

```python
#4. HOW MANY PEOPLE SURVIVED IN B5 CABIN
df=pd.read_csv('titanic.csv')
df.groupby('cabin')['survived'].sum().loc['B5']
```

```python
#5. COUNT OF FEMALES WHOSE AGE IS LESS THAN 20 AND SURVIVED
df.groupby(['sex','age','survived'])['survived'].sum().loc[('female',20,1)]
```

```python
#6. TOTAL FARE PAID BY EMBARKED S FEMALES
df.groupby(['embarked','sex'])['fare'].sum().loc[('S','female')]
```

```python
#7. COUNT OF PEOPLE IN PCLASS 3 HAVING ONLY 1 PERSON WITH THEM
df.groupby(['pclass','sibsp'])['name'].count().loc[(3,1)]
```

```python
#8. MAXIMUM FARE PAID IN EMBARKED Q
df.groupby('embarked')['fare'].max().loc[('Q')]
```

```python
#9. AVG TIME SPEND BY THE PEOPLE IN SUPPORT DEPARTMENT WHO GOT PROMOTED
df=pd.read_csv('hr_data.csv')
df.groupby(['promotion_last_5years','department'])['time_spend_company'].mean().loc[(1,'support')]
```

```python
#10. AVG NO OF PROJECTS PEOPLE WITH HIGH SALARY ARE HANDLING IN IT DEPT
df.groupby(['salary','department'])['number_project'].mean().loc[('high','IT')]
```

```python
#11. TOTAL NUMBER OF PROJECTS HANDLED BY SUPPORT DEPT
df.groupby('department')['number_project'].sum().loc[('support')]
```

```python
#12. COUNT OF PEOPLE IN SALES DEPT WHO HAVE WORK ACCIDENT
df.groupby(['department','Work_accident'])['Work_accident'].count().loc[('sales',1)]
```
